```python
import os
import openai
import gradio as gr
from pyChatGPT import ChatGPT
import warnings
from transformers import pipeline
```

```python
pipe = pipeline(model ='mallikrao2/asr_en_model') #import model from Hugging face
```

```python
pipe.device
```

```python
def transcribe(audio):
    text = pipe(audio)['text']
    return text

# interface
iface = gr.Interface(
    fn = transcribe,
    inputs = gr.Audio(source = 'microphone',type = 'filepath'),
    outputs ='text',
    title = 'ASR HUGGING FACE MODEL',
    description= 'Model',
)
```

```python
iface.launch()
```
